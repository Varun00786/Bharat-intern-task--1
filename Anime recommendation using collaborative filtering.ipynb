```python
ratings=ratings[ratings["user_id"].isin(y)]
ratings
```

```python
anime.drop(["genre","type","episodes","rating","members"],axis=1,inplace=True)
```

```python
main_df=ratings.merge(anime,on="anime_id")
main_df
```

```python
num_of_ratings=main_df.groupby("anime_id")["rating"].count().reset_index().rename(columns={"rating":"no_of_ratings"})
num_of_ratings
```

```python
final_df=main_df.merge(num_of_ratings,on="anime_id")
final_df
```

```python
final_df=final_df[final_df.no_of_ratings>=50]
final_df
```

```python
final_df.drop_duplicates(["user_id","anime_id"],inplace=True)
```

```python
pivot_df=final_df.pivot_table(columns="user_id",index="name",values="no_of_ratings")
pivot_df
```

```python
pivot_df.fillna(0,inplace=True)
```

```python
from scipy.sparse import csc_matrix
movie_sparse=csc_matrix(pivot_df)
```

```python
movie_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm="brute")
```

```python
model.fit(movie_sparse)
```

```python
distances,suggestions=model.kneighbors(pivot_df.iloc[1,:].values.reshape(1,-1),
                                       n_neighbors=6)
```

```python
distances
```

```python
suggestions
```

```python
def recommend_movie(movie_name):
    movie_id=np.where(pivot_df.index==movie_name)[0][0]
    distances,suggestions=model.kneighbors(pivot_df.iloc[movie_id,:].values.reshape(1,-1),n_neighbors=6)
    for i in range(len(suggestions)):
        if i==0:
            print("movies are")
        if not i:
            h=(pivot_df.index[suggestions[i]])
            for i in range(len(h)):
                print(h[i])
```

```python
recommend_movie("Kimi no Na wa.")
```
